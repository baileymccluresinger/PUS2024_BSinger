```python
import numpy as np
import scipy as sp
from scipy import stats
from scipy.stats import norm
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
np.random.seed(312)
#np.random.randn? #gives information and dimensions for the function
data = np.random.randn(200)
plt.hist(data, density=1)
xr = np.linspace(-3, 3, 100)
plt.plot(xr, norm.pdf(xr, loc=0, scale=1))
plt.xlabel("some data")
plt.ylabel("count/frequency")
```

```python
mean, stdev = sp.stats.norm.fit(data)
mean, stdev
```

```python
l = 1
data2 = np.random.poisson(l, 200)
plt.hist(data2, density=1)
xr2 = np.arange(0, 5, 1)
mean, stdev = sp.stats.norm.fit(data2)
plt.plot(xr2, norm.pdf(xr2, loc=mean, scale=stdev), label="norm")
plt.plot(xr2, poisson.pmf(xr2, mu=1))
plt.xlabel("some data")
plt.ylabel("count/frequency")
```

```python
l = 100
data = np.random.poisson(l, 200)
plt.hist(data, density=1)
xr = np.arange(60, 140, 1)
xr_continuous = np.linspace(60,140,100)
mean, stdev = sp.stats.norm.fit(data)
plt.plot(xr_continuous, norm.pdf(xr_continuous, loc=mean, scale=stdev), label="norm")
plt.plot(xr.astype(int), sp.stats.poisson.pmf(xr.astype(int), mu=l))
plt.xlabel("some data")
plt.ylabel("count/frequency")
```
